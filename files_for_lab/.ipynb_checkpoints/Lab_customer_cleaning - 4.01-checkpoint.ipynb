```python
# 7.Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. 
# Store continuous data into a continuous_df variable and do the same for discrete_df

numerical_df.nunique()
```

```python
def differentiate_variables(numerical_df):
    continuous_variables = []
    discrete_variables = []

    for column in numerical_df.columns:
        variable = numerical_df[column]
        if isinstance(variable[0], (int, float)):
            continuous_variables.append(column)
        else:
            discrete_variables.append(column)

    continuous_df = numerical_df[continuous_variables]
    discrete_df = numerical_df[discrete_variables]

    return continuous_df, discrete_df
```

```python
continuous_df, discrete_df = differentiate_variables(numerical_df)
```

```python
# 11.Look for outliers in the continuous variables. (HINT: Thereâ€™s a good plot to do that!)

def detect_outliers(continuous_df):
    Q1 = continuous_df.quantile(0.25)
    Q3 = continuous_df.quantile(0.75)
    IQR = Q3 - Q1

    outliers = ((continuous_df < (Q1 - 1.5 * IQR)) | (continuous_df > (Q3 + 1.5 * IQR))).any(axis=1)

    sns.boxplot(data=continuous_df[outliers])
    plt.title("Outliers in Continuous Variables")
    plt.xlabel("Variables")
    plt.ylabel("Values")
    plt.show()

detect_outliers(continuous_df)

```

```python
# 1.Open the categorica df that we created before.
cat_df = pd.read_csv('categorical.csv')
cat_df.head()
```

```python
# 2. Plot all the categorical variables with the proper plot. What can you see?
# Comments: I can't see a normal distribution between them. California is the state with the most insurance policies.
# The majority is living in the suburbs, are employees, have a lower education level, and trust more in the agent to do the work.
# are married and using a medium car.

def plot_continuous_variables(cat_df):
    for col in cat_df.columns:
        sns.histplot(data=cat_df, x=col, kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()

plot_continuous_variables(cat_df)
```

```python
# 3. There might be some columns that seem to be redundant, check their values to be sure. What should we do with them?
# Comments: I will keep all, but I will drop "Offer" column, include Bachelor, Master and Doctor in one category inside Education
# and I will join Two doors and four door car in one variable.  

def check_categorical_variables(cat_df):
    for col in cat_df.columns:
        unique_values = cat_df[col].unique()
        num_unique_values = len(unique_values)
        data_type = cat_df[col].dtype

        print(f"Column: {col}")
        print(f"Data Type: {data_type}")
        print(f"Unique Values: {unique_values}")
        print(f"Number of Unique Values: {num_unique_values}")
        print("----------------------")

check_categorical_variables(cat_df)
```

```python
#Dropping renew_offer_type
cat_df = cat_df.drop('renew_offer_type', axis=1)
cat_df
```

```python
#Remplacing Two-Door Car, Four-Door Car for just Car
def group_vehicle_class(vehicle_class):
    if vehicle_class in ['Two-Door Car', 'Four-Door Car']:
        return 'Car'
    else:
        return vehicle_class

cat_df['vehicle_class'] = cat_df['vehicle_class'].apply(group_vehicle_class)
display(cat_df['vehicle_class'])
```

```python
#Include Bachelor, Master and Doctor in one category inside Education
def group_education(education):
    if education in ['Bachelor','Master', 'Doctor']:
        return 'Bachelor or above'
    else:
        return education

cat_df['education'] = cat_df['education'].apply(group_education)
display(cat_df['education'])
```
