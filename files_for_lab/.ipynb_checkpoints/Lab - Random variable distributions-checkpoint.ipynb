```python
data_path = r'C:\Users\Andrea\Documents\DATA\lab-cleaning-numerical-data\files_for_lab\numerical1.csv'
data_num = pd.read_csv(data_path)
data_num.head()
```

# 2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
def distributions_plots(data_num):
    for col in data_num.columns:
        sns.histplot(data=data_num, x=col, kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()

distributions_plots(data_num)
```

# 3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
data_num.info()
```

```python
import numpy as np

data_num_log = np.log(data_num['customer_lifetime_value'])

plt.hist(data_num_log, bins=30)
plt.title('Distribution (Log-transform)')
plt.show()
```

```python
data_num_exp = np.exp(data_num['customer_lifetime_value'])

plt.hist(data_num_exp, bins=30)
plt.title('Distribution (Exponential)')
plt.show()
```

```python
import numpy as np
data_num_power = np.power(data_num['customer_lifetime_value'], 2)

plt.hist(data_num_power, bins=30)
plt.title('Distribution (Power)')
plt.show()
```

```python
from scipy import stats

data_num_boxcox, _ = stats.boxcox(data_num['customer_lifetime_value'])
plt.hist(data_num_boxcox, bins=30)
plt.title('Distribution (boxcox)')
plt.show()
```
